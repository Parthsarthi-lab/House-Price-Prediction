```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Loading the dataset

```python
df = pd.read_csv("BostonHousing.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Checking the missing values
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
sns.heatmap(data=df.corr())
```

```python
plt.scatter(df['CRIM'],df["MEDV"])
plt.xlabel("Crime Rate")
plt.ylabel("MEDV")
```

```python
# Independent and dependent features
X= df.drop(columns=["MEDV"])
y = df['MEDV']
```

```python
X.head()
```

```python
y.head()
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test= scaler.fit_transform(X_test)
```

```python
# Dumping 
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

#### What is a pickle file?

A pickle file in Python is a serialized binary file that allows you to save Python objects (such as data structures, models, or custom classes) to disk. Here’s why you might use pickle in a data science project:<br>

Object Serializatio<br>n:
Pickle allows you to serialize (convert to a byte stream) Python objec<br>ts.
Serialization is essential for saving objects to disk or transmitting them across netwo<br>rks.
Data Persist<br>ence:
In data science, you often work with large datasets, trained machine learning models, or complex data struc<br>tures.
Pickle lets you save these objects to disk, so you can load them later without re-computing or re-training.

##### Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# checking the parameters
regression.get_params()
```

```python
pred = regression.predict(X_test)
```

Checking the accuracy of our predictions

```python
sns.scatterplot(x=pred,y=y_test)
```

```python
error = y_test - pred
```

```python
error
```

```python
sns.histplot(error,kde=True)
```

```python
# Plot between residuals and prediction
plt.scatter(pred,error) # Model performs well if the plot has no pattern
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(np.sqrt(mean_squared_error(y_test,pred)))
```

```python
# R2 Score
from sklearn.metrics import r2_score
score=r2_score(y_test,pred)
print(score)
```

#### Dumping the Model

```python
pickle.dump(regression,open('model.pkl','wb')) # wb mode means writing a binary file
```
